```python
df = pd.read_csv('/kaggle/input/confer-extension-results/Part1/llama_part1_subset4_results.csv')
df = df.dropna()
df
```

```python
def get_type(row):

    uid = row['UID']
    trigger = row['trigger']

    if trigger == 'again':
        return 3

    if uid.startswith('type4'):
        return 1
    elif uid.startswith('type5p_possessive'):
        return 2   
    elif uid.startswith('type5a'):
        return 2  
```

```python
df['type'] = df.apply(get_type,axis=1)
df['type'].value_counts() 
```

```python
df['premise_type'].value_counts() 
```

```python
def is_original(row):
    
    premise_type = row['premise_type']
    if row['type'] in [1,2]:
        if premise_type == 'related':
            original = True
        else: 
            original = False
    else: # type 3 
        if premise_type == 'unrelated':
            original = True
        else: 
            original = False
    return original
```

```python
df['original'] = df.apply(is_original,axis=1)
df['original'].value_counts() 
```

```python
result = df.groupby(['type', 'original']).apply(compute_human_accuracy).reset_index(name='accuracy')
result.sort_values(by=['type', 'original'], ascending=[True, False], inplace=True)
result
```
